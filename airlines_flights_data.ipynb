# does ticket price change based on the departure time and arrival time?

```python
data.head()
```

```python
price_by_departure = data.groupby('departure_time')['price'].mean().sort_values(ascending=False)
print("Average Price by Departure Time:")
print(price_by_departure)
```

```python
price_by_arrival = data.groupby('arrival_time')['price'].mean().sort_values(ascending = True)
print("Average price by Arival Time:")
print(price_by_arrival)
```

```python
# Price by both departure and arrival time
price_by_dep_arr = data.groupby(['departure_time', 'arrival_time'])['price'].mean().unstack()
print("\nPrice by Departure and Arrival Time:")
print(price_by_dep_arr)

# Heatmap visualization
plt.figure(figsize=(12, 8))
sns.heatmap(price_by_dep_arr, annot=True, fmt='.0f', cmap='YlOrRd')
plt.title('Ticket Price Heatmap: Departure Time vs Arrival Time')
plt.show()
```

```python
sns.catplot(x = 'departure_time',y = 'price',kind = 'bar',data = data)

plt.show()
```

```python
sns.catplot(x = 'arrival_time',y = 'price',kind = 'bar',data = data)
plt.show()
```

```python
sns.relplot(x = 'arrival_time',y = 'price' , data = data , col = 'departure_time' , kind = 'line')
plt.show()
```
